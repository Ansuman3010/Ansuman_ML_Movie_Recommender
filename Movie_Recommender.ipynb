```python
movie["overview"] = movie["overview"].fillna(movie["title"])
movie["tagline"] = movie["tagline"].fillna(movie["title"])
movie.isnull().sum()
```

```python
import ast
def strolist(item):
    l=[]
    for i in ast.literal_eval(item):
        l.append(i['name'])
    return l

def cast(item):
    l=[]
    cnt=0;
    for i in ast.literal_eval(item):
        l.append(i['name'])
        cnt=cnt+1
        if(cnt>5):
          break;
    return l

def crew(item):
    l = []
    for i in ast.literal_eval(item):
        if (i['job'] == 'Director') or (i['job']=='Writer'):
            l.append(i['name'])
    return l
```

```python
movie['genres']=movie['genres'].apply(strolist)
movie['keywords']=movie['keywords'].apply(strolist)
movie['production_companies']=movie['production_companies'].apply(strolist)
movie['cast']=movie['cast'].apply(cast)
movie['crew']=movie['crew'].apply(crew)
```

```python
movie["genres"] = movie["genres"].apply(lambda x: " ".join(x))
movie["keywords"] = movie["keywords"].apply(lambda x: " ".join(x))
movie["cast"] = movie["cast"].apply(lambda x: " ".join(x))
movie["crew"] = movie["crew"].apply(lambda x: " ".join(x))
movie["production_companies"] = movie["production_companies"].apply(lambda x: " ".join(x))
```

```python
movie['tags']=movie['genres']*4+movie['keywords']*4+movie['production_companies']+movie['overview']+movie['cast']+movie['crew']+movie['tagline']
```

```python
v = movie['vote_count']
R = movie['vote_average']

C = movie['vote_average'].mean()
m = movie['vote_count'].quantile(0.90)   

movie['weighted_rating'] = (v/(v+m) * R) + (m/(v+m) * C)
```
